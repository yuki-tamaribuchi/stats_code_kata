```
import numpy as np
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```
spotify_df=pd.read_csv('spotify_datasets/genres_v2.csv')
```

```
spotify_df.head()
```

```
spotify_df.info()
```

```
spotify_df=spotify_df.drop(['song_name','Unnamed: 0','title','uri','track_href','analysis_url','type','id'],axis=1)
```

```
spotify_df.info()
```

```
key_dummies=pd.get_dummies(spotify_df['key'],prefix='key')
time_signature_dummies=pd.get_dummies(spotify_df['time_signature'],prefix='time_signature')

le_genre=LabelEncoder()
genre_labels=pd.DataFrame(le_genre.fit_transform(spotify_df['genre']))

spotify_df=spotify_df.drop(['key','time_signature','genre'],axis=1)
spotify_df=pd.concat([spotify_df,key_dummies,time_signature_dummies],axis=1)

scaler=StandardScaler()
scaled_spotify_df=scaler.fit_transform(spotify_df)
```
